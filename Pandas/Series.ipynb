### Merging Joining Concatination

```python
employees = pd.DataFrame({
    'employee_id': [1, 2, 3, 4],
    'Name': ['Mac', 'Kam', 'Sak', 'Bo'],
    'department': ['HR', 'IT', 'HR', 'IT']
})

# Salary information 
salaries = pd.DataFrame({
    'employee_id': [1, 2, 3, 6, 7],
    'Salary': [50000, 60000, 55000, 65000, 90000],
    'bonus': [5000, 6000, 5500, 6500, 4000]
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees, salaries, on='employee_id', how='inner')
```

```python
pd.merge(employees, salaries, on='employee_id', how='outer')
```

```python
pd.merge(employees, salaries, on='employee_id', how='left')
```

```python
pd.merge(employees, salaries, on='employee_id', how='right')
```

## Pivot Tables

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': [
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South',
        'East', 'West', 'North', 'South'
    ],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),
    'Rep': [
        'John', 'Mary', 'Bob', 'Alice',
        'John', 'Mary', 'Bob', 'Alice',
        'John', 'Mary', 'Bob', 'Alice',
        'John', 'Mary', 'Bob', 'Alice',
        'John', 'Mary', 'Bob', 'Alice'
    ]
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
df
```

```python
pd.pivot_table(df, values= 'Sales', index= 'Region', columns='Product')
```

```python
pivot = pd.pivot_table(df, values= ['Sales', 'Units'], index= 'Region', columns='Product')
pivot
```

```python
# Crosstab
pd.crosstab(df['Region'], df['Product'])
```
