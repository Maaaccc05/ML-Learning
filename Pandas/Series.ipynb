# Pandas Series

A series is a one dimentional labeled array capable of holding any data type. The axis lables are collectively called index

```python
labels = ['a', 'b', 'c']
list = [10, 20, 30]
arr = np.array([10,20,30])
dict = {1:10, 2:20, 3:30}
```

```python
pd.Series(list, index=labels)
```

```python
pd.Series(dict)
```

# DataFrames
When You combine multiple series together then it's called as Dataframes

```python
data = {
    'Name': ['Mac', 'Kam', 'Sak', 'Bo'],
    'Age': [21, 22, 20, 25],
    'City': ['In', 'In', 'Cam', 'Cam'],
    'Salary': [2000000, 30000, 40000, 50000]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
data_list = [
    ['Macc', 20, 'In', 20000000],
    ['Sak', 19, 'Cam', 3000000],
    ['Kam', 22, 'In', 900000],
    ['Bo', 25, 'Cam', 1000000]
]

df2 = pd.DataFrame(data_list)
columns = ["Name", "Age","City", "Salary"]
df2 = pd.DataFrame(data_list, columns=columns)
df2
```
