# DataFrames
When You combine multiple series together then it's called as Dataframes

Creating DataFrame

```python
data = {
    'Name': ['Mac', 'Kam', 'Sak', 'Bo'],
    'Age': [21, 22, 20, 25],
    'City': ['In', 'In', 'Cam', 'Cam'],
    'Salary': [2000000, 30000, 40000, 50000]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
data_list = [
    ['Macc', 20, 'In', 20000000],
    ['Sak', 19, 'Cam', 3000000],
    ['Kam', 22, 'In', 900000],
    ['Bo', 25, 'Cam', 1000000]
]

df2 = pd.DataFrame(data_list)
columns = ["Name", "Age","City", "Salary"]
df2 = pd.DataFrame(data_list, columns=columns)
df2
```

```python
df2
```

Selection and indexing in columns

```python
df2[['City','Name']]
```

Creating new Column


```python
df2["designation"] = ["doc", "eng", "doc", "eng"]
df2
```

### Missing Data

Finding Missing data

```python
data = {
    'A' : [1, 2, 3, 4, 5],
    'B' : [np.nan, 2, 3, 4, 5],
    'C' : [1, 2, 3, np.nan, np.nan],
    'D' : [1, np.nan, np.nan, np.nan, 5]
}

df3 = pd.DataFrame(data)
```

```python
# df3.isna().sum()
# df3.isna().any()
df3
```

Removing Missing Data

```python
df3.dropna()
```

```python
df3
```

```python
df3.dropna(thresh=4)
```

Filling the missing data

```python
df3.fillna(0)
```

```python
values = {'A': 0, 'B': 1, 'C': 2, 'D':3}
df3.fillna(value=values)
```

```python
df3.fillna(df3.mean())
```
