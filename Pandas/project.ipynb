```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'anime.csv')
```

```python
df.head()
```

```python
def extract_episode(txt):
    check = False
    data = ""
    for i in txt:
        if i == ")" :
            check = False
            return data
        if check == True:
            data = data + i
        if i == '(':
            check = True
```

```python
df['Title'].apply(extract_episode)
```
