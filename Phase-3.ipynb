### Advanced Operation with Business example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("=== Companay Data ===")
print("\n Data Shape ", sales_data.shape)
print("1st 3 restaurant", sales_data[:3]) # or you can enter like sales_data[0:3] also more detailed
print("\n1st 3 restaurant", sales_data[:, 1:]) 


```

### Total revenue of all restaurant

```python
print(np.sum(sales_data, axis=0))
yearly_rev = np.sum(sales_data[:, 1:], axis=0) #axis=0 calculate in vertical way despite of any order
print(yearly_rev)
print(np.sum(yearly_rev))

# mini sales 

min_sale = np.min(sales_data[:, 1:], axis=1) # axis= 1 only calculate for 1 particular line in horizontal way
print(min_sale)

# max sale
max_sale = np.max(sales_data[:, 1:], axis=0)
print(max_sale)

# avg sale per restaurant 
avg_sale = np.mean(sales_data[:, 1:], axis= 1)
print(avg_sale)
```

```python
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print(cumsum)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Avg cumsum per restaurant")
plt.xlabel("Years")
plt.ylabel("Income")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])
print("Vector addition ", vector1 + vector2)

print("Multiplication Of vector ", vector1 * vector2)
print("Dot product ",np.dot(vector1,vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2)))
print(angle)
```
