```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 5, 11)
x
y = x ** 2
y
```

```python
plt.plot(x,y)
plt.title("Graph")
plt.xlabel("Macc")
plt.ylabel("Kamm")
```

### Subplot in Matplotlib

```python
plt.subplot(2,2,2)
plt.plot(x ** 2, x ** 2)
plt.subplot(2,2,3)
plt.plot(x,y)
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(y ** 2, x)
```
