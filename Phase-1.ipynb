### List vs NumPy Array


```python
import numpy as np
import time

start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n List operation time = ", time.time() - start)

start = time.time()
py_npArray = np.arange(1000000) * 2
print("\n Numpy arrays operation time = ", time.time() - start)
```

### Creating Array from scratch

```python
import numpy as np
zeros = np.zeros((4, 5)) # creates matrix 4 rows 5 columns
print(zeros)

ones = np.ones((5, 4)) #5 rows 4 columns
print("Ones array: \n", ones) 

full = np.full((3, 4), 69)
print("Full array: \n", full)

random = np.random.random((3,3))
print("Random array: \n", random)

sequence = np.arange(0, 11, 3)
print("Sequence array: \n", sequence)
```
